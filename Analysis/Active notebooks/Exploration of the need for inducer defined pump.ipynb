## Pump specific speeds:

Experiments (not done by me) have shown that the actual pump efficiency does not significantly fall off with respect to speed, usually less than 2-3$\%$. For a given pump design the relationship between the fluid flow rate ($Q$), the developed head rise ($\Delta H$), and the rotational speed ($N$) can be defined by the so called 'Pump affinity laws' which are defined as follows:

(1) Pump volume flow rate varies directly with speed:

$$
\frac{Q_1}{Q_2} = \frac{N_1}{N_2}
$$

(2) Pump developed head rise varies with the square of the rotational speed:

$$
\frac{\Delta H_1}{\Delta H_2} = \frac{{N_1}^2}{{N_1}^2}
$$

(3) Pump driving power varies with the cube of the rotational speed:

$$
\frac{hp_1}{hp_2} = \frac{{N_1}^2}{{N_2}^2}
$$


Beginning with these parameters as defined by the motor requirements allows us to immediately derive the dimensionless pump specific speed ($N_s$) using the following equation:

$$
N_s = \frac{N(Q)^{1/2}}{\Delta H^{3/4}}
$$

Where:

 $N_s =$ Dimensionless suction specific speed  
 $N = $ Rotational speed of the pump [RPM]  
 $Q = $ Engine required flow rate [gpm]  
 $\Delta H =$ Required head rise [ft]  
 
 This method of calculation requires that we choose an operating rotational speed. However reasonable numbers have been discussed as a result of the previous calculations and the currently available COTS electric motors capable of delivering the expected power requirements as being somewhere in the range of about 7000-15000 RPM.  
**NOTE:** I will start at the lower end of these speeds and examine the effects of raising the rotational speed across the entire range of expected values.

## Design chriteria:
|         | density $\rho$ [kg/m^3] | Pump inlet suction pressure [psia] | pump discharge pressure [psia] | pump weight flowrate [lbm/sec] | pump rotational speed [RPM] |
|-------- | ----------------------- | ---------|-------|--------|----------|
| ethanol | 789                     | 14.7     | 380   | 1.412  | 7000
| LOX     | 1141                    | 14.7     | 380   | 1.836  | 7000


```python
#Bring in Fluid properties:
import math as m

#physical constants
g_0 = 9.81                                           # gravitational acceleration [m/s/s]
R = 8.3145                                           # Universal gas constant [Pa*m^3/K*mol]

# Propellant properties
lox = {'M': 31.999 / 1000,                           # Molar mass [kg/mol]
       'T_e': 90.2,                                  # Propellant temperature [K]
       'C_pl': 1.70 * 1000,                          # Liquid phase specific heat [J/kg*K]
       'C_pv': 0.91 * 1000,                          # Vapor phase secific heat (constant pressure) [J/kg*K]
       'h_v': 213.05 * 1000,                         # Heat of vaporization [J/kg]
       'T_v': 90.2,                                  # Vaporization temperature [K]
       'R_p': R / (31.999 / 1000),                   # Specific gas constant 
       'p_t': (350+70) * 1.1 * 6894,                 # Propellant tank initial pressure [Pa]
       'a': 1.382 / 10,                              # van der Wall constant A [Pa*m^6/mol^2]
       'b': 0.03186 / 1000,                          # van der Wall constant B [m^3/mol]
       'rho': 1141,                                  # Density [kg/m^3]
       'p_v': 101.4E3,                               # Operating point vapor pressure [Pa]
       'name': 'lox'
      }

ipa = {'M': 60.1 / 1000,                             # Molar mass [kg/mol]
       'T_e': 293,                                   # Propellant temperature [K]
       'C_pl': 7.25 * 1000 * (60.1 / 1000),          # Liquid phase specific heat [J/kg*K]
       'C_pv': 1.75 * 1000,                          # Vapor phase secific heat (constant pressure) [J/kg*K]
       'h_v': 36800 * (60.1 / 1000),                 # Heat of vaporization [J/kg]
       'T_v': 355.8,                                 # Vaporization temperature [K]
       'R_p': R / (60.1 / 1000),                     # Specific gas constant 
       'p_t': ((350                                  # Chamber pressure [psi]
             + 145                                   # Regen pressure drop [psi]
             + 70)                                   # Injector pressure drop [psi]
             * 1.1 * 6894),                          # Propellant tank initial pressure [Pa]
       'a': 12.18 / 10,                              # van der Wall constant A [Pa*m^6/mol^2]
       'b': 0.08407 / 1000,                          # van der Wall constant B [m^3/mol]
       'rho': 789,                                   # Density [kg/m^3]
       'p_v': 8.84E3,                                # Operating point vapor pressure [Pa]
       'name': 'ipa'
      }

# rocket model (assuming sea-level operation)
isp = 246 * 0.90                                     # specific impulse (sea-level, optimistic estimate) [s]
f = 3.2E3                                            # thrust [N]
mdot_t = f/(g_0 * isp)                               # total mass flowrate [kg/s]
chamber_p= 350 * 6.895 * 1000                        # chamber pressure (assumed to be 380 psi currently) [Pa] 
loss_factor = 1.15                                   # estimate of line and injector losses
OF = 1.3                                             # mixture ratio
mdot_o = mdot_t / (1 + (1/OF))                       # oxidizer mass flowrate [kg/s]
mdot_f = mdot_t / (1 + OF)                           # fuel mass flowrate [kg/s]
p_i = 101.3E3*3                                      # inlet pressure (currently 1 atm (gauge)) [Pa]
delta_p = chamber_p * loss_factor - p_i              # required pump discharge pressure [Pa]
rpm = 12000                                          # Estimated operational point [RPM]


def freedom_units(prop):
    rho = prop['rho'] * 0.062428
    h_v = prop['p_v'] * 0.000145038 * 144 / rho
    if prop == ipa:
        mdot = mdot_f * 2.20462
    elif prop == lox:
        mdot = mdot_o * 2.20462
    delta_psi = delta_p * 0.000145038                # total pressure rise [psia]   
    q = mdot * 449 / rho                             # volumetric flowrate [gpm]
    h_p = 144 * delta_p * 0.000145038 / rho          # required head rise [ft]
    h_i = p_i * 0.000145038 * 144 / rho              # inlet head [ft]
    npsh_a = h_i - h_v                               # Net Positive Suction Head (available) [ft] 
    n_s = rpm * (q**0.5) / (h_p**0.75)               # Pump specific speed [Dimensionless]
    return rho, h_v, mdot, q, h_p, npsh_a, delta_psi, n_s

print("required pressure head:")
print("%.3f" % freedom_units(ipa)[4], "ft (IPA)")
print("%.3f" % freedom_units(lox)[4], "ft (LOX)", "\n")
print("Developed pressure:")
print("%.3f" % freedom_units(ipa)[6], "psia (IPA)")
print("%.3f" % freedom_units(lox)[6], "psia (LOX)", "\n")
print("Mass flowrate:")
print("%.3f" % freedom_units(ipa)[3], "gpm (IPA)")
print("%.3f" % freedom_units(lox)[3], "gpm (LOX)", "\n")
print("Pump specific speed:")
print("%.3f" % freedom_units(ipa)[7], "(IPA)")
print("%.3f" % freedom_units(lox)[7], "(LOX)", "\n")
print("Net positive suction head available:")
print("%.3f" % freedom_units(ipa)[5], "(IPA)")
print("%.3f" % freedom_units(lox)[5], "(LOX)")
```

Critical net positive suction head [$NPSH_r$] is used to describe the minimum suction head above the propellant vapor pressure required to supress cavitation. this property is defined in relation to the pump specific speed as:

$$
N_{ss} = \frac{N(Q^{1/2})}{NPSH_r^{3/4}}
$$

In addition to the $NPSH_c$ at the pump operating point, $NPSH_r$ must be considered at all operating ranges to mitigate cavitation durring start up procedures.The starting $NPSH_r$ depends on the rate of acceleration and on the control system of the engine.

The other important property to characterise is the Thoma parameter [$\tau$] which is defined as:

$$
\tau = \frac{NPSH_r}{\Delta H} = (\frac{N_s}{N_{ss}})^{4/3}
$$

The thoma parameter is a function of design quality and specific speed. for a given vehicle, the suction characteristics ($N_{ss}$) determines the maximum permisible RPM at the given design flowrate. In general this means that a high $N_{ss}$ or a high $NPSH_a$ allows for a higher allowable RPM. It is desireable to operate the pump at the highest allowable speed.

```python
def suction_speed(prop):
    
    n_ss = (rpm * freedom_units(prop)[3]**0.5) / (freedom_units(prop)[5]**0.75)     # Pump suction specific speed [Dimensionless]
    tau = (freedom_units(prop)[7] / n_ss)**(4/3)                                    # Thoma parameter [Dimensionless]
    return n_ss, tau
    
print("Pump suction specific speed")
print("%.2f" % suction_speed(ipa)[0], "(IPA)")
print("%.2f" % suction_speed(lox)[0], "(LOX)", "\n")
print("Thoma parameter:")
print("%.2f" % suction_speed(ipa)[1], "(IPA)")
print("%.2f" % suction_speed(lox)[1], "(LOX)")
```
